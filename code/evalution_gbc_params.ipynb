```python
import pandas as pd
import seaborn as sns
import numpy
import matplotlib.pyplot as plt

sns.set_theme()
pd.set_option("display.max_colwidth", None)
```

```python
filename = "comparison"
results = pd.read_csv(f"../results/{filename}.csv")
```

```python
results = results[results["feature_selector"] == "SelectKBest"]
results = results[results["classifier"] == "GradientBoostingClassifier"]
results = results[results["n_features"] == 4]

results["score"] = results.apply(
    lambda x: 10_000 * x.accuracy - 200 * x.n_features, axis=1
)


results.sort_values("accuracy_top_20pc", ascending=False).head(10)
```
